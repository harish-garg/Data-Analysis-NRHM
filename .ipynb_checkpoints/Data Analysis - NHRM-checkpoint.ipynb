```python
#!/usr/bin/python

# Import the necessary libraries

import pandas as pd
```

```python
# Import a sample xls file into a pandas data strcuture
file = "DHIS (APRIL 15-MARCH 16)/BHIWAPUR BLOCK/PHC/Copy of MonthlyProgressReportMIES_PHC Jawali_Apr-2015.xls"

xl = pd.ExcelFile(file)

df = xl.parse('Sheet 1')
```

```python
# Cleanup the DataFrame

## Columns
### Drop the column which has no values and no column header.

df.drop('Unnamed: 0', axis=1, inplace=True)

### Rename the columns to something more descriptive.
df.columns = [
                'Code', 'Indicator', 'Type', 'Apr-2015', 'May-2015', 'Jun-2015',
                'Jul-2015', 'Aug-2015', 'Sep-2015', 'Oct-2015', 'Nov-2015',
                'Dec-2015', 'Jan-2016', 'Feb-2016', 'Mar-2016', 'Total'
             ]
## Rows
### Delete the rows relating to Category Part D which has no values and were hidden in the file
part_d = df[df.Code == 'Part D'].index.tolist()[0]
part_e = df[df.Code == 'Part E'].index.tolist()[0]
df.drop(df[df.index.isin(range(part_d, part_e))].index.tolist(), inplace=True)

### Delete the rows which have no values(for examples the categpry headers)
df.drop(df[df['Type'].isnull()].index.tolist(), inplace=True)


'''
part_df = df.Code[df.Code.str.contains('^Part', na=False)]
part_df_index = df[df.Code.str.contains('^Part', na=False)].index.tolist()
#print part_df
#print part_df_index
#print len(part_df_index)
df['cat1'] = ''
n = 0
for i in part_df_index:
    #print i, part_df[i]
    
    if n < len(part_df_index) - 1:
        df['cat1'][df.index.isin(range(i, part_df_index[n+1]))] = part_df[i]
        #print i, part_df_index[n+1]
    else:
        df['cat1'][df.index.isin(range(i, df.tail(1).index.tolist()[0]+1))] = part_df[i]
        #print i, df.tail(1).index.tolist()[0]
    n = n + 1
'''
```

```python
# Add a new column for top level category
df['Category Level 1'] = ''

## Add Category REPRODUCTIVE AND CHILD HEALTH

df['Category Level 1'][df.Code.str.contains('^\d\.', na=False)] = 'REPRODUCTIVE AND CHILD HEALTH'
df['Category Level 1'][df.Code.str.contains('^1[0-2]\.', na=False)] = 'REPRODUCTIVE AND CHILD HEALTH'

## Add category Other Programmes
df['Category Level 1'][df.Code.str.contains('^13\.', na=False)] = 'Other Programmes'

## Add category Health Facility Services
df['Category Level 1'][df.Code.str.contains('^1[4-5]\.', na=False)] = 'Health Facility Services'

## Mortality Details
df['Category Level 1'][df.Code.str.contains('^17\.', na=False)] = 'Mortality Details'
```

```python
# Add a new column for 2nd level category
df['Category Level 2'] = ''
df['Category Level 2'][df.Code.str.contains('^1\.', na=False)] = 'Ante Natal Care Services ANC'
df['Category Level 2'][df.Code.str.contains('^2\.', na=False)] = 'Deliveries'
df['Category Level 2'][df.Code.str.contains('^3\.', na=False)] = 'Number of Caesarean C-Section deliveries performed at '
df['Category Level 2'][df.Code.str.contains('^4\.', na=False)] = 'Pregnancy outcome & weight of new-born'
df['Category Level 2'][df.Code.str.contains('^5\.', na=False)] = 'Complicated pregnanices'
df['Category Level 2'][df.Code.str.contains('^6\.', na=False)] = 'Post - Natal Care'
df['Category Level 2'][df.Code.str.contains('^7\.', na=False)] = 'Medical Termination of Pregnancy (MTP)'
df['Category Level 2'][df.Code.str.contains('^8\.', na=False)] = 'RTI/STI Cases'
df['Category Level 2'][df.Code.str.contains('^9\.', na=False)] = 'Family Planning'
df['Category Level 2'][df.Code.str.contains('^10\.', na=False)] = 'CHILD IMMUNIZATION'
df['Category Level 2'][df.Code.str.contains('^11\.', na=False)] = 'Number of Vitamin A doses '
df['Category Level 2'][df.Code.str.contains('^12\.', na=False)] = 'Number of cases of Childhood Diseases reported during the month0-5 years: '
df['Category Level 2'][df.Code.str.contains('^13\.', na=False)] = 'Blindness Control Programme'
df['Category Level 2'][df.Code.str.contains('^14\.', na=False)] = 'Patient Services'
df['Category Level 2'][df.Code.str.contains('^15\.', na=False)] = 'Laboratory Testing'
df['Category Level 2'][df.Code.str.contains('^17\.', na=False)] = 'Details of deaths reported during the month with probable causes:'
```
